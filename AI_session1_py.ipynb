```
#!pip install matplotlib
#loading matplotlib library
import matplotlib.pyplot as plt
```

```
x=[2,9,3,7]
y=[7,15,5,9]
plt.xlabel("Distance")# creating x axis
plt.ylabel("Time")#creating y axis
plt.plot(x,y,label="railway track",marker='*',color='red') #this will draw line
plt.scatter(x,y,s=200,label="railwar station",color='black') # s is the size to increase size of particular 
plt.bar(x,y,label='height of railway station') # to draw bar charts
plt.grid(color='green')
plt.legend() # it will show label in the plot
plt.show()#draw graoh
```
